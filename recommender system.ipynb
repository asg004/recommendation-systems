```python
brand_names = review_df.set_index('id')['categories'].to_dict()
n_dimension = len(sales_df.about_product.unique())
n_keys = len(sales_df.rating.unique())
print("Number of unique brands:", n_dimension)
print("Number of unique keyss:", n_keys)
print("The full rating matrix will have:", n_dimension*n_keys, 'elements.')
print('----------')
print("Number of sales:", len(sales_df))
print("Therefore: ", len(sales_df) / (n_dimension*n_keys) * 100, '% of the matrix is filled.')
```

```python
import torch
import numpy as np
from torch.autograd import Variable
from tqdm import tqdm_notebook as tqdm

class MatrixFactorization(torch.nn.Module):
    def __init__(self, n_dimension, n_keys, n_factors=20):
        super().__init__()

        self.dimension_factors = torch.nn.Embedding(n_dimension, n_factors)

        self.keys_factors = torch.nn.Embedding(n_keys, n_factors)
        self.dimension_factors.weight.data.uniform_(0, 0.05)
        self.keys_factors.weight.data.uniform_(0, 0.05)

    def forward(self, data):

        dn_dimension, n_keys = data[:,0], data[:,1]
        return (self.dimension_factors(dn_dimension)*self.keys_factors(n_keys)).sum(1)


    def predict(self, dimension, keys):
        return self.forward(dimension, keys)
```

```python

from torch.utils.data.dataset import Dataset
from torch.utils.data import DataLoader


class Loader(Dataset):
    def __init__(self):
        self.sales = sales_df.copy()


        self.sales['about_product'] = pd.to_numeric(self.sales['about_product'])
        self.sales['rating'] = pd.to_numeric(self.sales['rating'])


        dimensions = sales_df.about_product.unique()
        review = sales_df.rating.unique()




        self.product_id2idx = {o:i for i,o in enumerate(dimensions)}
        self.rating2idx = {o:i for i,o in enumerate(review)}


        self.idx2product_id = {i:o for o,i in self.product_id2idx.items()}
        self.idx2rating = {i:o for o,i in self.rating2idx.items()}


        self.sales.rating = sales_df.rating.apply(lambda x: self.rating2idx[x])
        self.sales.about_product = sales_df.about_product.apply(lambda x: self.product_id2idx[x])


        self.x = self.sales.drop(['rating', 'about_product'], axis=1).values
        self.y = self.sales['rating'].values
        self.x, self.y = torch.tensor(self.x), torch.tensor(self.y)
```

```python
num_epochs = 128
cuda = torch.cuda.is_available()

print("Is running on GPU:", cuda)

model = MatrixFactorization(n_dimension, n_keys, n_factors=8)
print(model)
for name, param in model.named_parameters():
    if param.requires_grad:
        print(name, param.data)

if cuda:
    model = model.cuda()


loss_fn = torch.nn.MSELoss()

optimizer = torch.optim.Adam(model.parameters(), lr=1e-3)


def self():
  self.sales['about_product'] = pd.to_numeric(self.sales['about_product'].str.replace(',', ''))
  train_set = Loader()
def train_loader():
  train_loader = DataLoader(train_set, 128, shuffle=True)
```

```python
def tqdm():
  for it in tqdm(range(num_epochs)):
    losses = []
    for x, y in train_loader:
        if torch.cuda.is_available():
            x, y = x.cuda(), y.cuda()
        optimizer.zero_grad()
        outputs = model(x)
        loss = loss_fn(outputs.squeeze(), y.type(torch.float32))
        losses.append(loss.item())
        loss.backward()
        optimizer.step()
    if len(losses) > 0:
        print("iter #{}".format(it), "Loss:", sum(losses) / len(losses))
    else:
        print("iter #{}".format(it), "Loss: N/A")
```
